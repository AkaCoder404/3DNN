# Visualizing ModelNet40 Normal Resampled

```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset_path = "data/modelnet40_normal_resampled"
airplane_example_file = dataset_path + "/airplane/airplane_0001.txt"

# Load file
point_set = [line.rstrip().split(',')[0:3] for line in open(airplane_example_file)]
point_set = [[float(dim) for dim in point] for point in point_set]
point_set = np.array(point_set)

X, Y, Z = point_set[:, 0], point_set[:, 1], point_set[:, 2]

plt.scatter(X, Y, Z)
```

```python
# First N Points
n = 2048
X, Y, Z = point_set[:n, 0], point_set[:n, 1], point_set[:n, 2]
plt.scatter(X, Y, Z)
```

```python
# Plot normalized
from custom_datasets import pc_normalize
point_set = pc_normalize(point_set)
X, Y, Z = point_set[:, 0], point_set[:, 1], point_set[:, 2]
plt.scatter(X, Y, Z)
```
