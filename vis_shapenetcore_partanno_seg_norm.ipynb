# Visiualizing ShapeNetCore Segmentation
Dataset from https://shapenet.cs.stanford.edu/ericyi/shapenetcore_partanno_v0.zip

```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset_path = "data/shapenetcore_partanno_segmentation_benchmark_v0_normal"
categories = {line.rsplit()[0]: line.rsplit()[1] for line in open(dataset_path + "/synsetoffset2category.txt")}
print(categories)

airplane_example_file_list = os.listdir(dataset_path + "/" + categories["Airplane"])
rand_int = np.random.randint(len(airplane_example_file_list))
airplane_example_file = dataset_path + "/" + categories["Airplane"] + "/" + airplane_example_file_list[rand_int]

# Load file
point_set = [line.rstrip().split(' ')[0:7] for line in open(airplane_example_file)]
point_set = [[float(dim) for dim in point] for point in point_set]
point_set = np.array(point_set)

X, Y, Z = point_set[:, 0], point_set[:, 1], point_set[:, 2]
C = point_set[:, 6]

print("Total points: ", len(point_set))

colormap = plt.cm.get_cmap('coolwarm')
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X, Y, Z, c=C, cmap=colormap)
```

```python
# Load file
n = 1024 # Points
X, Y, Z = point_set[:n, 0], point_set[:n, 1], point_set[:n, 2]
C = point_set[:n, 6]

colormap = plt.cm.get_cmap('coolwarm')
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X, Y, Z, c=C, cmap=colormap)
```
